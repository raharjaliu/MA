```python
import pandas as pd
from matplotlib import pyplot as pt

%matplotlib inline
```

```python
df = pd.read_csv('../results/40-47.csv', header=-1)
df.columns = ['POSITION', 'TREE_ID', 'DIV_TIME', 'TTD1', 'TTD2']
```

```python
df1 = df[(df.POSITION <= 47) & (df.POSITION >= 40)]
```

```python
pt.scatter(df1.DIV_TIME, df1.TTD1)
pt.scatter(df1.DIV_TIME, df1.TTD2)
pt.title('Division time vs Time-to-death (position 40-47)')
pt.xlabel('Division time (seconds)')
pt.ylabel('Time-to-deatg (seconds)')

pt.gcf().set_size_inches(6, 6)
```

```python
df2 = df[(df.POSITION <= 55) & (df.POSITION >= 48)]
```

```python
pt.scatter(df2.DIV_TIME, df2.TTD1)
pt.scatter(df2.DIV_TIME, df2.TTD2)
pt.title('Division time vs Time-to-death (position 48-55)')
pt.xlabel('Division time (seconds)')
pt.ylabel('Time-to-deatg (seconds)')

pt.gcf().set_size_inches(6, 6)
```
