```python
import os
import re
import time

import pandas as pd
```

```python
re.match()
```

```python
file = open("../results/unsyn_log.txt", "r")
log_unsyn = file.readlines()
file.close()
```

```python
positions = []
valid_no_div = []
valid_one_div = []
```

```python
positions = list(filter(lambda x: "Processing position" in x, log_unsyn))
positions = [re.search("[0-9]+", x).group(0) for x in positions]
positions = [int(n) for n in positions]
```

```python
valid_no_div = list(filter(lambda x: "singly-placed trees with 0 div" in x, log_unsyn))
valid_no_div = [re.search("trees with 0 div: ([0-9]+)", x).group(1) for x in valid_no_div]
valid_no_div = [int(n) for n in valid_no_div]
```

```python
valid_one_div = list(filter(lambda x: "singly-placed trees with 1 div" in x, log_unsyn))
valid_one_div = [re.search("trees with 1 div: ([0-9]+)", x).group(1) for x in valid_one_div]
valid_one_div = [int(n) for n in valid_one_div]
```

```python
stats = pd.DataFrame({'POSITION':positions,
              'VALID_ONE_DIV': valid_one_div,
              'VALID_NO_DIV': valid_no_div})
```

```python
stats
```

```python
print("Total:\n\tNO_DIV:%d\n\tONE_DIV:%d" % (stats.VALID_NO_DIV.sum(), stats.VALID_ONE_DIV.sum()))
```
