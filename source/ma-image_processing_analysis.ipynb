```python
import cv2
import os
import numpy as np

from matplotlib import pyplot as pt
%matplotlib inline
```

```python
testpic = cv2.imread("D:/MA/test/eli-new-unsync-bf-41/caspase/caspasexy41c1.tif")
```

```python
testpic = cv2.imread("D:/MA/test/eli-new-unsync-bf-41/in-focus/before/bf_in-focusxy41c1c1.tif")
cv2.imwrite("D:/workspace/msc-thesis-manu/images/pos_41_in_t0.png", testpic)
```

```python
# test the picture's encoding
testpic.dtype
```

```python
# test the picture's shape
testpic.shape
```

```python
# translate image to intensity image
testpic_br = .299 * testpic[:,:,0] + .587 * testpic[:,:,1] + .114 * testpic[:,:,2]
```

```python
# plot the picture
pt.imshow(testpic)
pt.gcf().set_size_inches(5, 5)
```

```python
## plot intensity distribution

pt.hist(testpic_br.ravel(), color="black", bins=20, normed=True)
pt.gcf().set_size_inches(4, 3)
pt.ticklabel_format(style="sci")
pt.ylabel("probability")
pt.xlabel("Intensity (8-bit)")
pt.tight_layout()
pt.savefig("D:/workspace/msc-thesis-manu/images/pos_41_in_t0_hist.pdf")
```

```python
## increase brightness by 100

testpic_br_c = testpic_br.astype(np.uint16) + 100
testpic_br_c[testpic_br_c >= 255] = 255
testpic_br_c = testpic_br_c.astype(np.uint8)

pt.imshow(testpic_br_c)
pt.gcf().set_size_inches(5, 5)
cv2.imwrite("D:/workspace/msc-thesis-manu/images/pos_41_in_t0_br_up100.png", testpic_br_c)
```

```python
## plot intensity distribution

pt.hist(testpic_br_c.ravel(), color="black", bins=20, normed=True)
pt.gcf().set_size_inches(4, 3)
pt.ticklabel_format(style="sci")
pt.ylabel("probability")
pt.xlabel("Intensity (8-bit)")
pt.tight_layout()
pt.savefig("D:/workspace/msc-thesis-manu/images/pos_41_in_t0_br_up100_hist.pdf")
```
